```python
from qiskit import QuantumCircuit
from qiskit_aer import Aer

# Create a quantum circuit with 1 qubit and 1 classical bit
qc = QuantumCircuit(1, 1)

# Put qubit into superposition
qc.h(0)

# Measure the qubit
qc.measure(0, 0)

# Get simulator backend
backend = Aer.get_backend('qasm_simulator')

# Run the circuit (1 shot = 1 random bit)
job = backend.run(qc, shots=1)
result = job.result()

# Get output
counts = result.get_counts()
print("Quantum Random Bit:", counts)

# Draw the circuit
qc.draw()
```

```python
# Run the same circuit multiple times
job = backend.run(qc, shots=100)
result = job.result()

counts = result.get_counts()
print("Quantum Random Bits Distribution:", counts)
```

```python
from qiskit.visualization import plot_histogram

plot_histogram(counts)
```

```python
def quantum_random_bit():
    job = backend.run(qc, shots=1)
    result = job.result().get_counts()
    return int(list(result.keys())[0])

# Generate 10 random bits
for i in range(10):
    print(quantum_random_bit())
```
